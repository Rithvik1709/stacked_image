```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

folder = "images/" 

image_files = [os.path.join(folder, f) for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
if not image_files:
    raise ValueError("No images found in the folder!")
images = [cv2.imread(file).astype(np.float32) for file in image_files]
base_shape = images[0].shape[:2] 
images = [cv2.resize(img, (base_shape[1], base_shape[0])) for img in images]
stacked = np.mean(images, axis=0).astype(np.uint8)
output_path = "stacked_output.jpg"
cv2.imwrite(output_path, stacked)
print(f" Stacked image saved as '{output_path}'")
cv2.imshow("Stacked Image", stacked)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(cv2.cvtColor(stacked, cv2.COLOR_BGR2RGB))
plt.title("Stacked Image (Average)")
plt.axis("off")
plt.show()

```
